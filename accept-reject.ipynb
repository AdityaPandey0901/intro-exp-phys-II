```python
def p(x):
    return np.sin(x-0.5)**2 * np.exp(-x**2) + .2*np.exp(-x**2)
```

```python
x = np.linspace(-3,3,100)
y = p(x)
```

```python
   
def attempt():
    xmin = -3
    xmax = 3
    pmax = 0.8
    t = (xmax-xmin)*np.random.rand() + xmin
    y = np.random.rand()
    if y < p(t)/ pmax:
        return t
    else:
        return False
    
def accept_reject(N):
    n_accept=0
    x_list = []
    while n_accept < N:
        x = attempt()
        if x:
            n_accept = n_accept+1
            x_list.append(x)
    return x_list
            
```

```python
x = accept_reject(10000)
ugly_stuff = plt.hist(x, bins=100)
```
